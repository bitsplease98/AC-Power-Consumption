```python
def hours_to_timing(hour):
    if hour in [22, 23, 0, 1, 2, 3]:
        timing = 'Night'
    elif hour in range(4, 12):
        timing = 'Morning'
    elif hour in range(12, 17):
        timing = 'Afternoon'
    elif hour in range(17, 22):
        timing = 'Evening'
    
    return timing
```

# Webscraping

Scraping Gurgaon Temperature for the month of September and August for the year 2019 from worldweather website

```python
from selenium import webdriver
import re
import time
```

```python
driver=webdriver.Edge('msedgedriver.exe')
driver.maximize_window()
driver.get("https://www.worldweatheronline.com/gurgaon-weather-history/haryana/in.aspx")
```

```python
mintemp=[]
maxtemp=[]
daylist=[]
monthlist=[]
```

```python
for month in ["08","09"]:
    gmonth=int(month)
    if month=="08":
        days=31
    if month=="09":
        days=30
    for day in range(1,days+1):
        gday=day
        day=str(day)
        if len(day)==1:
            day="0"+day
        date=day+"-"+month+"-2019"
        try:
            #Sending the date
            driver.find_element_by_xpath("//input[@name='ctl00$MainContentHolder$txtPastDate']").send_keys(date)
            time.sleep(2)
            #Clicking on Submit Button
            driver.find_element_by_xpath('//input[@name="ctl00$MainContentHolder$butShowPastWeather"]').click()
            
            #Getting the Minimum and Maximum Temperature for the day
            temp=driver.find_element_by_xpath("//div[@class='days-collapse-temp']").text

            Min=temp.split("\n")[1].split("/")[0]
            Max=temp.split("\n")[1].split("/")[1]

            Min = int(re.sub("\D", "", Min))
            Max = int(re.sub("\D", "", Max))
        except:
            Min=np.NaN
            Max=np.NaN
        daylist.append(gday)
        monthlist.append(gmonth)
        mintemp.append(Min)
        maxtemp.append(Max)
```

```python
data=pd.DataFrame()
data['Day']=daylist
data['Month']=monthlist
data['Mintemp']=mintemp
data['Maxtemp']=maxtemp
data
```

```python
#Saving the scraped Data
data.to_csv('gurgaontemp.csv')
```

### Removing Outliers

```python
from scipy.stats import zscore
```

```python
#finding optimum threshold in the range 3 to 5
def threshold():
    for i in np.arange(3,5,0.2):
        data=df.copy()
        data=data[(z<i).all(axis=1)]
        loss=(df.shape[0]-data.shape[0])/df.shape[0]*100
        print('With threshold {} data loss is {}%'.format(np.round(i,1),np.round(loss,2))) 
```

```python
z=np.abs(zscore(df))
threshold()
```

```python
#We use threshold as 3.2 because data is expensive and upon evaluation of models this threhold turned out to be the best.
df=df[(z<3.2).all(axis=1)]
```

# Modelling Phase

Importing necessary libraries

```python
from sklearn.model_selection import train_test_split,cross_val_score
```

```python
#importing models
from sklearn.neighbors import KNeighborsRegressor
from sklearn.linear_model import LinearRegression,Lasso,Ridge,ElasticNet
from sklearn.svm import SVR
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor,AdaBoostRegressor,GradientBoostingRegressor
from xgboost import XGBRegressor
```

```python
from sklearn.metrics import r2_score,mean_absolute_error,mean_squared_error
```

```python
#Creating list of models and another list mapped to their names
models=[KNeighborsRegressor(),LinearRegression(),Lasso(),Ridge(),DecisionTreeRegressor(),
       RandomForestRegressor(),AdaBoostRegressor(),GradientBoostingRegressor(),XGBRegressor()]

names=['KNeighborsRegressor','LinearRegression','Lasso','Ridge','DecisionTreeRegressor',
       'RandomForestRegressor','AdaBoostRegressor','GradientBoostingRegressor','XGBRegressor']
```

```python
def createmodels(model_list,xtrain,xtest,ytrain,ytest,n):
    name=[]
    meanabs=[]
    meansqd=[]
    rootmeansqd=[]
    r2=[]
    
    #Creating models
    for i,model in enumerate(model_list):
        model.fit(xtrain,ytrain)
        p=model.predict(xtest)
        
        #Calculating scores of the model and appending them to a list
        name.append(n[i])
        meanabs.append(np.round(mean_absolute_error(p,ytest),4))
        meansqd.append(np.round(mean_squared_error(p,ytest),4))
        rootmeansqd.append(np.round(np.sqrt(mean_squared_error(p,ytest)),4))
        r2.append(np.round(r2_score(p,ytest),2))
    
    #Creating Dataframe
    data=pd.DataFrame()
    data['Model']=name
    data['Mean Absolute Error']=meanabs
    data['Mean Squared Error']=meansqd
    data['Root Mean Squared Error']=rootmeansqd
    data['R2 Score']=r2
    data.set_index('Model',inplace = True)
    return data
```

```python
#Spliting the data into test and train in 1:4 ratio
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.25,random_state=7)
```

```python
print('---------------------------------------Performance--------------------------------------------')
createmodels(models,xtrain,xtest,ytrain,ytest,names)
```

From Above Linear Regressor gives overwhelming performance followed by XGBRegressor, Gradient Boost and Random Forest. Hence we will perform hyperparameter tuning on them
